# **MakeTenJudgement**

## **概要**
make ten と呼ばれることもあるゲームで、多くの場合四桁程度の数字を四則演算、入れ替えを用いて10を作ることを目指すゲームを解くためのプログラムです。

(例1)　8530　→　8 + 5 - 3 + 0 = 10

(例2)　3451　→　((4 - 3) + 1) * 5

のように変換し、作成可能かどうかを返します。

## **操作方法**

まず最初にモードを半角数字1か2で選択します。

1→四桁の数字を与え、その数字が10を作成可能かどうか判別します。

(四桁以上の数字を与えると下四桁を抽出します。)


2→三桁の数字を与え、その数字に0~9のどの数字を加えても10を作成可能か判別します。

次に判断したい数字を入力します。

(三桁以上の数字を与えると下三桁を抽出します。)

(例)　1を入力後5830を入力　→　10が作成可能であるというメッセージが出て終了
## **制作者のコメント**

最初、入れ替えという概念に苦戦し、四桁の数字を入れ替え、演算子を入れ替え、計算順序を入れ替えといったことをすべてfor文で行おうとし、ネストは深くすごく長いコードが出来上がりました。

しかし自身でも読みづらさを感じていたため、一度書いていたコードをすべて見直し、考え直したところ、入れ替えのライブラリはきっといくらでもあるだろうということに気づき探したところ案の定すぐ見つかったためそのライブラリを用いました。

すると10を作成可能か判断する関数がすぐ完成し、もうちょっと早くライブラリの存在を思い出せたらよかったと思いました。

```python
'''コード処理メモ
- 4桁の数字の入力(int型四桁)を受け取る
- 各桁の数字を抽出し、NumPy配列に格納(←この段階で一桁の数字を四つ持つlistを作成)
- itertoolsを用いて四桁の数字の組み合わせ出す
- itertoolsかなんかで文字列として四則とカッコの組み合わせを出す
- 文字列をevalで計算
- その数字で10を作成できるか判定
'''

%reset -f
import numpy as np
import sys
import itertools


def extract_digits(input_number,number_of_digits):# 入力された数字から各桁の数字を抽出し、NumPy配列に格納する関数
  # NumPy配列をゼロで初期化
  output_number_ = np.zeros(4, dtype=int)
  for i in range(number_of_digits):
    # 入力された数字から桁を抽出
    output_number_[(number_of_digits - 1) - i] = (input_number // (10 ** i)) % 10
  return output_number_


def is_possible_to_make_ten(*four_numbers):#  # 4つの数字で10を作成できるか判定する関数
    operators = ['+', '-', '*', '/']
    for numbers in itertools.permutations(four_numbers):
        for ops in itertools.product(operators, repeat=3):
            expression = f"({numbers[0]} {ops[0]} {numbers[1]}) {ops[1]} ({numbers[2]} {ops[2]} {numbers[3]})"
            try:
                if eval(expression) >= 10 - 1e-9 and eval(expression) <= 10 + 1e-9:
                    return True
            except ZeroDivisionError:
                continue

            expression = f"({numbers[0]} {ops[0]} ({numbers[1]} {ops[1]} {numbers[2]})) {ops[2]} {numbers[3]}"
            try:
                if eval(expression) >= 10 - 1e-9 and eval(expression) <= 10 + 1e-9:
                    return True
            except ZeroDivisionError:
                continue

            expression = f"{numbers[0]} {ops[0]} ({numbers[1]} {ops[1]} ({numbers[2]} {ops[2]} {numbers[3]}))"
            try:
                if eval(expression) >= 10 - 1e-9 and eval(expression) <= 10 + 1e-9:
                    return True
            except ZeroDivisionError:
                continue

            expression = f"(({numbers[0]} {ops[0]} {numbers[1]}) {ops[1]} {numbers[2]}) {ops[2]} {numbers[3]}"
            try:
                if eval(expression) >= 10 - 1e-9 and eval(expression) <= 10 + 1e-9:
                    return True
            except ZeroDivisionError:
                continue

            expression = f"{numbers[0]} {ops[0]} (({numbers[1]} {ops[1]} {numbers[2]}) {ops[2]} {numbers[3]})"
            try:
                if eval(expression) >= 10 - 1e-9 and eval(expression) <= 10 + 1e-9:
                    return True
            except ZeroDivisionError:
                continue
    return False



##▽以下処理層

ignore_amount = 0##指示を無視した回数
mode_choice = int(input("モードを選択してください。  その数字が任意の数字を作成可能か判定する場合は1を、その数字が1～9のどの数字を入れても10が作成可能か調べる場合は2を押してください。"))


# ##初回で話を聞かなかった人への四回までの慈悲(例外処理)
for i in range(5):
  if mode_choice != 1 and mode_choice != 2:
    if ignore_amount < 4 :
      print("ちゃんと話聞いてね。")
      mode_choice = int(input("モードを選択してください。四桁の数字が10を作成可能か判定する場合は1を、三桁の数字が1～9のどの数字を入れても10が作成可能か調べる場合は2を押してください。"))
      ignore_amount +=1
    else:
      print("もういいです。")
      sys.exit("指示無視")


if mode_choice == 1:
  print("四桁の数字が10を作成可能か判定します。")
  number = int(input("4桁の数字を入力してください。  四桁以上の場合は下4桁を使用します。"))
  four_number = extract_digits(number,4)
  if is_possible_to_make_ten(*four_number) :
    print(f"{four_number}では10を作成できます。")
  else:
    # sys.exit(f"{four_number}では10を作成できません。")
    print(f"{four_number}では10を作成できません。")

elif mode_choice == 2:

  print("三桁の数字が1～9のどの数字を入れても10が作成可能か調べます。")
  number = int(input("3桁の数字を入力してください。  三桁以上の場合は下3桁を使用します。"))
  three_number = extract_digits(number,3)  # 最初の3桁を抽出
  for i in range(0, 10):  # 1から9までの数字を試す
    if is_possible_to_make_ten(*three_number, i):
      print(f"{three_number}に{i}を追加すると、10を作成できます。")
    else:
      print(f"{three_number}に{i}を追加しても、10は作成できません。")
```
